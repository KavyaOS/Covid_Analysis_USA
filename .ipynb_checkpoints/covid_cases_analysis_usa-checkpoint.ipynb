```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import datetime
```

```python
covid_cases_table = pd.read_csv('covid_data_usa/covid_confirmed_usafacts.csv')
covid_deaths_table = pd.read_csv('covid_data_usa/covid_deaths_usafacts.csv')
covid_population_table = pd.read_csv('covid_data_usa/covid_county_population_usafacts.csv')
```

```python
def get_weekly_data_separated_for_counties(dataframe):
    weekly_counts = dataframe.loc[:, '2020-01-26'::7] #getting total cases till that weekend
    weekly_counts = weekly_counts.diff(axis=1) #to change cumulative data to per week data
    weekly_counts = weekly_counts.drop(['2020-01-26'],axis=1) #to get rid of partial week count of first weekend
    weekly_counts.columns = pd.to_datetime(weekly_counts.columns) #to change datatype of dates from string to datetime
    weekly_counts.columns = weekly_counts.columns - datetime.timedelta(days=6) #To make weeks represent through monday dates instead of Sunday
    return weekly_counts
```

# Q1. Plot the weekly number of new cases of covid in the following states: California, Oregon, Washington, and Nevada. A week starts on Monday and ends on Sunday.

```python
'''Dropping all the unwanted columns'''
statewise_cases_table = covid_cases_table.drop(['countyFIPS','StateFIPS','County Name'], axis=1)

'''Dropping unrequired states'''
states = ['CA', 'OR', 'WA', 'NV']
statewise_cases_table = statewise_cases_table[statewise_cases_table['State'].isin(states)]

statewise_cases_table = statewise_cases_table.groupby(['State']).sum().reset_index()
states = statewise_cases_table['State']
weekly_counts = statewise_cases_table.loc[:, '2020-01-26'::7] #getting total cases till that weekend
weekly_counts = weekly_counts.diff(axis=1) #to change cumulative data to per week data
weekly_counts = weekly_counts.drop(['2020-01-26'],axis=1) #to get rid of partial week count of first weekend
weekly_counts.columns = pd.to_datetime(weekly_counts.columns) #to change datatype of dates from string to datetime

final_table = pd.merge(states,weekly_counts,left_index=True, right_index=True)
final_table = final_table.set_index('State')
final_table = final_table.transpose() #just to make it compatible to plot
final_table.index = final_table.index - datetime.timedelta(days=6) #To make weeks represent through monday dates instead of Sunday
final_table = final_table.rename_axis("Week")
display(final_table.head(5))
display(final_table.tail(5))
final_table.plot()
```

# Q2. Do those states follow the the same trends?

```python
'''To check the trend of each state by plotting individually'''
x_coordinates = final_table.index
y_coordinates_CA = final_table.iloc[:,0]
y_coordinates_OR = final_table.iloc[:,1]
y_coordinates_WA = final_table.iloc[:,2]
y_coordinates_NV = final_table.iloc[:,3]
figure = plt.figure(figsize=(20,6))
axes1 = figure.add_subplot(2, 2, 1)
axes1.set_title('Covid cases in California')
axes2 = figure.add_subplot(2, 2, 2)
axes2.set_title('Covid cases in Oregon')
axes3 = figure.add_subplot(2, 2, 3)
axes3.set_title('Covid cases in Washington')
axes4 = figure.add_subplot(2, 2, 4)
axes4.set_title('Covid cases in Nevada')
axes1.plot(x_coordinates,y_coordinates_CA)
axes2.plot(x_coordinates,y_coordinates_OR)
axes3.plot(x_coordinates,y_coordinates_WA)
axes4.plot(x_coordinates,y_coordinates_NV)
```

By seeing above graphs we can conclude that the states California, Oregon, Washington and Nevada follow same trends

# Q3.Plot the weekly number of covid deaths in the following states: California, Oregon, Washington,and Nevada

```python
'''Dropping all the unwanted columns'''
statewise_death_table = covid_deaths_table.drop(['countyFIPS','StateFIPS','County Name'], axis=1)

'''Dropping unrequired states'''
states = ['CA', 'OR', 'WA', 'NV']
statewise_death_table = statewise_death_table[statewise_death_table['State'].isin(states)]

statewise_death_table = statewise_death_table.groupby(['State']).sum().reset_index()
states = statewise_death_table['State']
weekly_counts = statewise_death_table.loc[:, '2020-01-26'::7] #getting total cases till that weekend
weekly_counts = weekly_counts.diff(axis=1) #to change cumulative data to per week data
weekly_counts = weekly_counts.drop(['2020-01-26'],axis=1) #to get rid of partial week count of first weekend
weekly_counts.columns = pd.to_datetime(weekly_counts.columns) #to change datatype of dates from string to datetime
final_table = pd.merge(states,weekly_counts,left_index=True, right_index=True)
final_table = final_table.set_index('State')
final_table = final_table.transpose()
final_table.index = final_table.index - datetime.timedelta(days=6) #To make weeks represent through monday dates instead of Sunday
final_table = final_table.rename_axis("Week")
display(final_table.head(5))
final_table.plot()
display(final_table.tail(5))
```

# Q4.How do the deaths and the new cases compare?

```python
'''To check the trend of deaths due to covid in each of those 4 states by plotting individually'''
x_coordinates = final_table.index
y_coordinates_CA = final_table.iloc[:,0]
y_coordinates_OR = final_table.iloc[:,1]
y_coordinates_WA = final_table.iloc[:,2]
y_coordinates_NV = final_table.iloc[:,3]
figure = plt.figure(figsize=(20,6))
axes1 = figure.add_subplot(2, 2, 1)
axes1.set_title('Covid cases in California')
axes2 = figure.add_subplot(2, 2, 2)
axes2.set_title('Covid cases in Oregon')
axes3 = figure.add_subplot(2, 2, 3)
axes3.set_title('Covid cases in Washington')
axes4 = figure.add_subplot(2, 2, 4)
axes4.set_title('Covid cases in Nevada')
axes1.plot(x_coordinates,y_coordinates_CA)
axes2.plot(x_coordinates,y_coordinates_OR)
axes3.plot(x_coordinates,y_coordinates_WA)
axes4.plot(x_coordinates,y_coordinates_NV)
```

By seeing output of question2 and question4 it is apparent that the deaths increased as the number of new covid cases increased in each state. But in initial stages deaths were more in Nevada and Oregon even when cases were less. 

# Q5.Compute the number of new covid cases per 100,000 population in each county per week. What are the 20 highest rates achieved. Show the date, county and the rate.

```python
'''Modifying data as required'''
countywise_cases_table = covid_cases_table.drop(['StateFIPS','State','County Name'], axis=1)
countywise_cases_table = countywise_cases_table.drop(countywise_cases_table[countywise_cases_table['countyFIPS']==0].index)
countywise_cases_table = countywise_cases_table.set_index('countyFIPS')
countywise_cases_table = get_weekly_data_separated_for_counties(countywise_cases_table)
countywise_cases_table.columns = countywise_cases_table.columns.strftime('%Y/%m/%d')
county_population_table = covid_population_table.set_index('countyFIPS')

countywise_cases_table = pd.merge(county_population_table, countywise_cases_table,on='countyFIPS')
countywise_cases_table.reset_index(inplace=True)

'''To calculate rate per 100000 population'''
countywise_cases_table.loc[:,'2020/01/27':] = (countywise_cases_table.loc[:,'2020/01/27':].mul(100000)).div(countywise_cases_table['population'], axis = 0)
    
result = countywise_cases_table.melt(id_vars=['countyFIPS','County Name','State','population'],value_name='rate',var_name='Week Date')
result = result.drop(['State','population'],axis=1) #filtering result to show what's required
result = result.nlargest(20,'rate').reset_index(drop=True)
display(result)
```

# Q6.Compute the number of covid deaths per 100,000 population in each county per week. What are the 20 highest rates achieved. Show the date, county and the rate.

```python
countywise_deaths_table = covid_deaths_table.drop(['StateFIPS','State','County Name'], axis=1)
countywise_deaths_table = countywise_deaths_table.drop(countywise_deaths_table[countywise_deaths_table['countyFIPS']==0].index)
countywise_deaths_table = countywise_deaths_table.set_index('countyFIPS')
countywise_deaths_table = get_weekly_data_separated_for_counties(countywise_deaths_table)
countywise_deaths_table.columns = countywise_deaths_table.columns.strftime('%Y/%m/%d')

county_population_table = covid_population_table.set_index('countyFIPS')

countywise_deaths_table = pd.merge(county_population_table, countywise_deaths_table,on='countyFIPS')
countywise_deaths_table.reset_index(inplace=True)

countywise_deaths_table.loc[:,'2020/01/27':] = (countywise_deaths_table.loc[:,'2020/01/27':].mul(100000)).div(countywise_deaths_table['population'], axis = 0)

countywise_deaths_table = countywise_deaths_table.melt(id_vars=['countyFIPS','County Name','State','population'],value_name='rate',var_name='Week Date')
countywise_deaths_table = countywise_deaths_table.drop(['State','population'],axis=1)
result = countywise_deaths_table.nlargest(20,'rate').reset_index(drop=True)
display(result)
```

There is a county called Wayne county which appeared in both, but those are from different states (with different countyFIPS). # Q8.Find the week that each county achieved their highest rate of new covid cases per 100,000 population. If a county reach the peak multiple ties pick the earliest one. Plot via a bar chart the number of counties that reached their peak each week. What does the plot indicate about the pandemic?

```python
'''To drop columns that are not needed'''
county_columns = countywise_cases_table[['countyFIPS','County Name']]
rates_table = countywise_cases_table.drop(['countyFIPS','County Name', 'State', 'population'],axis=1)

dates = rates_table.columns #retaining weekly dates for the purpose of reindexing later
max_values = pd.DataFrame() #empty dataframe to hold date and rate
max_values['rate'] = rates_table.max(axis=1)
max_values['date'] = rates_table.idxmax(axis=1)
result = county_columns.merge(max_values,left_index=True, right_index=True) #Table with highest value achieved by each county, rate and date
output_8 = max_values.groupby('date').count()
output_8 = output_8.reindex(dates, fill_value=0)
output_8 = output_8.rename_axis("Week")
display(output_8.head(5))
display(output_8.tail(5))
output_8.plot.bar(figsize=(20,8), grid=True)
```

Plot indicates that most of the counties reached max rate of covid cases after (2020/11/02) November 2020. Gradually it is tending to decrease in February 2021.

# Q9.Compute the total number of covid cases in each county and the number of covid deaths. Compute the percent of the population in each county that is still alive and has had covid. Show the top 20 rates. Is any county getting close to herd immunity, that is 70% or more people immune?

```python
countywise_cases_table = covid_cases_table[['countyFIPS','2021-02-07']]
countywise_cases_table = countywise_cases_table.rename(columns={"2021-02-07":"Total_Cases"})
countywise_deaths_table = covid_deaths_table[['countyFIPS','2021-02-07']]
countywise_deaths_table = countywise_deaths_table.rename(columns={"2021-02-07":"Total_Deaths"})
countywise_population_table = covid_population_table[['countyFIPS','County Name','population']]
countywise_population_table = countywise_population_table.drop(countywise_population_table[countywise_population_table['countyFIPS']==0].index)

stats_table = countywise_population_table.merge(countywise_cases_table.merge(countywise_deaths_table,on='countyFIPS'),on='countyFIPS')
stats_table['%_of_ppl_with_antibodies_to_covid'] = ((stats_table['Total_Cases'] - stats_table['Total_Deaths']) / stats_table['population']) * 100
display(stats_table.nlargest(20,'%_of_ppl_with_antibodies_to_covid').reset_index(drop=True))
```

As seen in the "%_of_ppl_with_antibodies_to_covid" column in the above table, highest percentage of people with antibodies to covid is approximately 30% among all counties, which clearly shows that none of the counties reached herd immunity.
